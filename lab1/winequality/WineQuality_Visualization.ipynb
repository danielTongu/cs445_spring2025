# Wine Quality Data Analysis

This notebook visualizes data from the UCI Wine Quality dataset.  
The dataset contains physicochemical properties of red wine samples and their associated quality ratings (scored between 0 and 10).  
Our goal is to create various plots using Matplotlib to explore and interpret aspects of the data.

**Source**: [Wine Quality Dataset - UCI Repository](https://archive.ics.uci.edu/ml/datasets/Wine+Quality)

```python
import pandas as pd
import matplotlib.pyplot as plt

def load_wine_data(csv_path):
    """
    Loads the red wine quality dataset from a CSV file.

    @param csv_path: Path to the CSV file.
    @return: DataFrame containing the wine data.
    """
    df = pd.read_csv(csv_path, sep=';')
    return df

# Load and preview the dataset
wine_data = load_wine_data('winequality/winequality-red.csv')

# Display the first few rows of the dataset
print("Wine Dataset Preview:")
print(wine_data.head())
```

## Bar Chart: Wine Quality Distribution

This bar chart shows how many wine samples received each quality score.  
It helps us understand how the wines are distributed in terms of sensory ratings.

```python
def plot_quality_distribution_bar_chart(dataframe):
    """
    Creates a bar chart showing the distribution of wine quality ratings.

    @param dataframe: DataFrame containing the wine data.
    """
    quality_counts = dataframe['quality'].value_counts().sort_index()

    plt.figure(figsize=(10, 6))
    plt.bar(quality_counts.index, quality_counts.values)

    plt.xlabel("Wine Quality Score", fontsize=12)
    plt.ylabel("Number of Samples", fontsize=12)
    plt.title("Wine Quality Distribution (Red Wine)", fontsize=14)

    #plt.savefig("DanielTongu_WineQualityBarChart.png")
    plt.show()

# Call the function to plot and save the bar chart
plot_quality_distribution_bar_chart(wine_data)
```

## Line Graph: Alcohol Content vs Wine Quality

This line graph shows the relationship between the alcohol content and the average wine quality score.  
We calculate the average quality score for each unique alcohol level. This helps us understand if wines with higher alcohol tend to be rated better.

```python
def plot_alcohol_vs_quality_line_graph(dataframe):
    """
    Creates a line graph that shows the relationship between alcohol content
    and the average wine quality score.

    @param dataframe: DataFrame containing the wine data.
    """
    # Grouping by alcohol level and computing mean quality
    grouped_data = dataframe.groupby('alcohol')['quality'].mean()

    plt.figure(figsize=(10, 6))
    plt.plot(grouped_data.index, grouped_data.values, marker='o')

    plt.xlabel("Alcohol Content (%)", fontsize=12)
    plt.ylabel("Average Wine Quality Score", fontsize=12)
    plt.title("Alcohol Content vs Wine Quality", fontsize=14)

    #plt.savefig("DanielTongu_AlcoholVsQualityLineGraph.png")
    plt.show()

# Call the function to plot and save the line graph
plot_alcohol_vs_quality_line_graph(wine_data)
```

## Pie Chart: Proportion of Wine Quality Ratings

This pie chart visualizes the proportion of different wine quality ratings in the dataset.  
It gives us a sense of how the wine quality scores are distributed as percentages.

```python
def plot_quality_pie_chart(dataframe):
    """
    Creates a pie chart of the proportion of wine samples by quality rating.

    @param dataframe: DataFrame containing the wine data.
    """
    quality_counts = dataframe['quality'].value_counts().sort_index()

    plt.figure(figsize=(8, 8))
    plt.pie(quality_counts.values, labels=quality_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title("Wine Quality Rating Proportions", fontsize=14)
    #plt.savefig("DanielTongu_WineQualityPieChart.png")
    plt.show()

# Call the function to plot and save the pie chart
plot_quality_pie_chart(wine_data)
```

## Box Plot: Distribution of Residual Sugar

This box chart displays the distribution of residual sugar in the red wine samples.  
It helps identify the spread and presence of any potential outliers in the sugar content.

```python
def plot_residual_sugar_box_chart(dataframe):
    """
    Creates a box plot to show the distribution of residual sugar in the dataset.

    @param dataframe: DataFrame containing the wine data.
    """
    plt.figure(figsize=(8, 6))
    plt.boxplot(dataframe['residual sugar'])

    plt.ylabel("Residual Sugar (g/dm^3)", fontsize=12)
    plt.title("Distribution of Residual Sugar in Red Wine", fontsize=14)
    #plt.savefig("DanielTongu_ResidualSugarBoxChart.png")
    plt.show()

# Call the function to plot and save the box chart
plot_residual_sugar_box_chart(wine_data)
```
